# 1. Main page
Extracting titles and url of respective single pages

```python
site = "https://www.whatmobile.com.pk/20000_to_30001_Mobiles"
#hdr = {'User-Agent': 'Mozilla/5.0'}
bookpage = requests.get(site)
soup = BeautifulSoup(bookpage.text, "html.parser")
print(soup.prettify())
```

```python
soup.find_all('td')
```

```python
soup.find_all('td')
```

```python

titles = soup.find_all('td', class_ ='BiggerText')
titles
```

```python
for i in titles:
    print(i.text.strip(), i.a['href'])
```

```python
title_db = []
url_db = []
for i in titles:
    name = i.text.strip()
    url = i.a['href']
    title_db.append(name)
    url_db.append(url)
```
